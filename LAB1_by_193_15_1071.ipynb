```python
from pyspark.sql.types import *

labels=[
  ("decile1b",IntegerType()),
  ("decile3",IntegerType()),
  ("ID", IntegerType()),
  ("decile1", IntegerType()),
  ("sex", IntegerType())
]

schema = StructType([StructField (x[0], x[1], True) for x in labels])
schema
```

```python
df.select(df.decile1b).show()
df.select(df.sex).show(truncate=True)
```

```python
from pyspark.sql.functions import col
df.select(col('sex')).show(truncate=False)
```

```python
df.select(df['sex']).show(truncate=False)
```

```python
from pyspark.sql.functions import lit,concat
df=df.withColumn('ID & sex',concat(col('ID'), lit(' '), col('sex')))
#df=df.withColumn('first_name',lit("New")).withColumn('second_column',concat(col('ID'), lit(' '), col('sex')))

df.show()
```

```python
df=df.withColumnRenamed('ID','id')
df.show()
```

```python
df.collect()
```

```python
df.groupBy('decile1').count().show()
df.groupBy('race').count().show()
df.groupBy('sex','grad').count().show()
```
